```python
# "M" correspond à un million de donwloads et "K" 1000.
def change_Size(size : str) -> int: 
    size = str(size) #Permet de s'assurer que même si on rentre un entier, le code fonctionne et renvoie l'entier sans modif
    size = size.upper() #Mettre tout sur la même base donc M et K en majuscule.
    if "M" in size:
        size =size.replace("M","") #on retire le M de l'écriture
        size = float(size)
        size = size * 1000000 #Mais on multiplie qd même la valeur renseignée par 1millions
    elif "K" in size: #Pareil pour K 
        size =size.replace("K","")
        size = float(size)
        size = size * 1000
    size = int(float(size))  
    return size    
```

```python
df["Size"] = df["Size"].apply(change_Size)
df
```
